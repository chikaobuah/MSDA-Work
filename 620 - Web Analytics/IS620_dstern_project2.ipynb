For this assignment, I revisited the data I used in Project 1. This is a large two-node dataset of campaign contributions from residents of New York State to candidates in the 2016 Presidential election. There are only edges between the node sets and not within them.

I downloaded this data as a csv file from the most recent quarterly report on the website for the Federal Elections Commission. I then split the data into two csv files - one for nodes (all contributors and candidates, deduplicated) and one for edges (all campaign contributions).

```python
import os
import csv
import networkx as nx
import matplotlib.pylab as plt
import math
from networkx.algorithms import bipartite as bi
%matplotlib inline
```

```python
reader = csv.reader(open("NYedges.csv", 'rU'), dialect=csv.excel_tab)
```

```python
G = nx.Graph()
```

```python
politicians = []
for row in reader:
    row = row[0].split(',')
    contributor = row[0]
    politician = row[1]
    G.add_edge(contributor, politician)
```

```python
print "Nodes: %d" %nx.number_of_nodes(G)
print "Edges: %d" %nx.number_of_edges(G)
```

```python
reader2 = csv.reader(open("candidates.csv", 'rU'), dialect=csv.excel_tab)
reader3 = csv.reader(open("contributors.csv", 'rU'), dialect=csv.excel_tab)
candidates = []
contributors = []
for row in reader2:
    candidates.append(row)
for row in reader3:
    contributors.append(row)
candidates = [item for each in candidates for item in each]
contributors = [item for each in contributors for item in each]
```

Now we will project the bipartite graph onto the candidate nodes. We see here that there is a core of "establishment" candidates with many bipartisan links. We can see some distinction between candidates on the left and right by the relative weight of the edges. There are a few "fringe" candidates with few links to other candidates.

```python
pols = bi.weighted_projected_graph(G, candidates, ratio=False) 
polWeights=[edata['weight'] for f,t,edata in pols.edges(data=True)] 
plt.figure(figsize = (15,10))
nx.draw_networkx(pols,width=polWeights, edge_color=polWeights)
```

Now we will look at a histogram of the weights to see how to best trim the nodes. We will use the trim_edges function from "Social Network Analysis for Startups" to trim edges at a weight of 2. 

```python
plt.hist(polWeights)
```

```python
def trim_edges(g, weight=1): 
    g2=nx.Graph()
    for f, to, edata in g.edges(data=True): 
        if edata['weight'] > weight:
            g2.add_edge(f,to,edata)
    return g2
```

```python
pols2 = trim_edges(pols, weight=2)
plt.figure(figsize = (10,10))
weights2 = [edata['weight'] for f,t,edata in pols2.edges(data=True)]
nx.draw_networkx(pols2, width=weights)
```

After reducing the size of the network, we can see that Democratic and Republican politicans are now divided into two distinct groups. (There is an exception in George Pataki, a moderate Republican governor of New York, although we would expect him to have quite a few bipartisan donors in the NY data set.) What's interesting about this reduced network is that we can infer where money might flow as candidates dwindle. If this data is representative of a national campaign contributions, we might expect all Democratic contributors to eventually line up behind Hillary Clinton, and likewise all Republican contributors to back Marco Rubio.
